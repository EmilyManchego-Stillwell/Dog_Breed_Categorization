```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
#Creating an empty list to hold the dog breed data as a list of lists
breed_data = []

#Looping through the list of dog breeds
for breed_folder in os.listdir('../Images'):
    if '-' not in breed_folder:
        continue
    
    #Splitting breed id and name from each folder in Images folder
    breed_id, breed_name = breed_folder.split('-', 1)
    #Verifying split worked correctly
    #print(f'breed_id:{breed_id}, breed_name:{breed_name}')
    
    #Pulling the image file from each breed folder and appending all information to the breed_data list above
    for image_file in os.listdir(f'../Images/{breed_folder}'):
        breed_data.append([breed_id, breed_name, image_file])

breed_data        
```

```python
#Making a dataframe out of the breed_data
breed_data_df = pd.DataFrame(breed_data, columns=['Breed_id', 'Breed_name', 'Img_file'])

breed_data_df.head()
```

```python
out_path = Path('../Data/Final_Project_Breed_data.csv')
breed_data_df.to_csv(out_path, index=False)
```
